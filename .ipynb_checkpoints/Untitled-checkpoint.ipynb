```python
import pandas as pd;
df = pd.read_csv('books.csv')


```

```python
# how many records
# cols
print(df.describe());
print(df.columns);
# df.count() df.dtypes()
```

```python
# which books has the most reviews, average rating of the book

df.sort_values('ratings_count', ascending = False).head(1)[['title', 'average_rating']]
# print(qq)
```

```python
# which has the highest average rating
max_value = df['average_rating'].max()
df[df['average_rating']==max_value][['title', 'average_rating']]
```
